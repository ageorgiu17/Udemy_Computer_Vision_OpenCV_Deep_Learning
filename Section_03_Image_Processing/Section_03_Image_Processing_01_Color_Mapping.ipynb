## Color Mapping

Am lucrat până acuma doar cu culorile create de către RGB. Această schemă de culori este una destul de veche, iar ca o la o laternativă pentru această schemă au apărut HSL(Hue-Saturation-Lightness) și HSV(Hue-Saturation-Value). Aceste scheme de culori sunt mai apropiate de modul în care oamenii percep culorile. Deși cele mai multe operațiuni o să le facem pe imagini ce folosesc RGB, nu strică să știm cum putem să schimbăm o imagine din RGB în HSL sau HSV, poate la un moment dat o să avem nevoie să facem asta. Să vedem prima dată ce diferențe există între RGB și restul. 

<img src='../SS/Section_03_01.png'>

Modelul de RGB reprezintă o combinație de culori între Red Green și Blue. Putem să afișăm acest concept ca și un cub (trei dimensiuni), unde lungimea, lățimea și înălțimea reprezintă câte una dintre culorile RGB, iar valorile sunt cuprinse între 0 și 255 pentru fiecare dimensiune. Schema de culori HSL sau HSV putem să le vizualizăm ca și un cilindru.

<img src='../SS/section_03_02.png'>

Pentru acestă schemă, de-a lungul cilindrului o să fie culori diferite (verde, galben, albastru, mov, etc.), iar acesta reprezintă partea de Hue. Partea de Saturation se referă la cât de mult din acel Hue o să avem pentru o anumită culoare. Partea de Lightness reprezintă cât de luminată sau de închisă este culoarea respectivă. Lecția respecitvă o să fie mai mult o recapitulare pentru metoda 'cv2.cvtColor()' deoarece pe parcursul acestui curs o să lucrăm doar cu imagini RGB

```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading the image
image = cv2.imread('../DATA/00-puppy.jpg')
```

```python
# showing the image
plt.imshow(image)
```

Am învățat că acele color channel sunt în ordinea diferită pentru OpenCV și pentru Matplotlib. OpenCV le are în ordinea Blue-Green-Red, iar Matplotlib le are în ordinea Red-Green-Blue. Pentru a afișa imaginea cu culorile corecte trebuie să facem transformarea din BGR în RGB. Asta se face după cum bine știm cu metoda 'cv2.cvtColor()'

```python
# inverting color channels
good_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# showing the image
plt.imshow(good_image)
```

La fel cum am făcut transformarea din BGR în RGB, tot așa putem să facem transformarea și din RGB în HLS, utilizând tot comanda 'cv2.cvtColor()'

```python
# changing from RGB to HSL
hsl_image = cv2.cvtColor(good_image, cv2.COLOR_RGB2HLS)

# showing the image
plt.imshow(hsl_image)
```

Imaginea respectivă arată ciudat deoarece Matplotlib așteaptă RGB ca și color channnel, din acest motiv sunt afișate așa. La fel putem schima din orice color channel în altul, OpenCV cam are toate conversile pregătite, trebuie doar să le găsim și să le apelăm.

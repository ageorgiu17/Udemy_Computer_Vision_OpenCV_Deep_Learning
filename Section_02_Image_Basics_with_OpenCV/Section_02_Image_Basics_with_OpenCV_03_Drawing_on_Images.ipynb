# Drawing on Images

În cadrul acestei lecții o să trecem peste procedeele prin care putem să desenăm anumite obiecte pe o imagine.

```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

Pentru partea practică în cadrul acestei lecții o să creem o imagine care nu are nimic pe ea, adică o imagine complet neagră. Pentru a crea o astfel de imagine putem să ne folosim de metoda numpy.zeros(). Metoda respectivă creează un array de date prin pe care după cum bine știm îl putem folosi ca și o imagine.

```python
blank_image = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
blank_image.shape
```

Mai sus am creat o imagine de 512 pe 512 pixeli cu 3 color channels. Putem să afișăm imaginea aceasta, dar aceasta o să fie neagră complet deoarece valorile din acel array sunt toate 0

```python
plt.imshow(blank_image)
```

În continuare o să începem să desenăm un dreptunghi (rectangle) pe această imagine. Pentru a desena o să ne folosim de metoda cv2.rectangle(). Metoda respectivă ia următoarele argumente:

    img = imaginea pe care să desenăm

    pt1 = tuplă care conține coordonatele punctului din stânga-sus al dreptunghiului

    pt2 = tuplă care conține coordinatele punctului din dreapta-jos ale dreptunghiului

    color = tuplă ce conține culoarea dreptunghiului sub formă de cod RGB

    thickness = lățimea linei sub formă de pixel size



De menționat este faptul că această metodă nu returnează acel shape, ci îl creează automat pe imagine, deci o să fie permanent pe imaginea pe care o oferim ca și input

```python
cv2.rectangle(img=blank_image, pt1=(300, 0), pt2=(510, 200), color=(0, 255, 0), thickness=8)
```

```python
plt.imshow(blank_image)
```

În continuare o să mai creem un pătrat localizat în mijlocul imaginii respective pentru a ne mai obișnui cu acest procedeu de a desena astfel de forme pe o imagine

```python
cv2.rectangle(img=blank_image, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=4)
```

```python
plt.imshow(blank_image)
```

Pentru a desena cercuri este un proces destul de similar, o să folosim metoda cv2.circle() căreia trebuie din nou să îi oferim imaginea pe care să deseneze acel cerc, culoarea și grosimea liniei cercului, diferența este la punctele de localizare. Nu o să îi mai oferim un set de două puncte, ci o să îi oferim centru cercului și raza acestuia. Centrul i se specifică cu ajutorul argumentului 'center' căruia îi oferim o tuplă de două valori, coordonatele pentru x și y, iar raza se specifică cu parametrul 'radius' și se exprimă în distanța pixelilor de la centrul cercului până la limita acestuia (o singură valoare)

```python
cv2.circle(img=blank_image, center=(100, 100), radius=50, color=(255, 0, 0), thickness=4)
plt.imshow(blank_image)
```

În cazul în care dorim să creem o formă, iar interiorul acestei forme să fie completat cu culoarea pe care o alegem, putem să facem asta prin specificarea valorii -1 pentru argumentul 'thickness'

```python
cv2.circle(img=blank_image, center=(200, 400), radius=50, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_image)
```

Pentru a crea o linie pe o imagine o să ne folosim de metoda cv2.line(), iar aceasta are nevoie de aceeași parametrii ca și metoda cv2.rectangle

```python
cv2.line(img=blank_image, pt1=(0, 0), pt2=(500, 500), color=(102, 255, 255), thickness=2)
plt.imshow(blank_image)
```

```python
cv2.line(img=blank_image, pt1=(200, 400), pt2=(400, 400), color=(102, 255, 255), thickness=2)
plt.imshow(blank_image)
```

## Recapitulare

În cadrul acestei lecții am învățat următoarele:

    1. Cum să ne creem o imagine care să fie complet neagră

        blank_image = np.zeros(shape=(512, 512, 3), dtype=np.int16)

    2. Cum să desenăm un dreptunghi pe o imagine

        cv2.rectangle(img=blank_image, pt1=(200, 200), pt2=(300, 300), color=(0, 255, 0), thickness=4)

    3. Cum să desenăm un cerc pe o imagine

        cv2.circle(img=blank_image, center=(100, 100), radius=50, color=(255, 0, 0), thickness=-1)

            # thickness=1 o să ne umple toată forma cu culoarea pe care am specificat-0

    4. Cum să desenăm o linie pe o imagine

        cv2.line(img=blank_image, pt1=(200, 400), pt2=(400, 400), color=(102, 255, 255), thickness=2)
    
    Metodele de mai sus nu retunează nimic, ele modifică imaginea originală. Dacă dorim ca imaginea să fie nemodificată trebuie fie să o recitim fie să creem o copie a aceste imagini.

# 1. Open the 'dog_backpack.png' image and dispaly it

```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading the image
image = cv2.imread('../DATA/dog_backpack.png')
```

```python
# displaying the image
plt.imshow(image)
```

```python
# Corectting the collor channel orders
good_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# displaying the new image
plt.imshow(good_image)
```

# 2. Flip the image on the vertical axes

```python
# flipping the image
flipped_image = cv2.flip(good_image, 0)

# showing the image
plt.imshow(flipped_image)
```

# 3. Draw an empty rectangle on the dogs face

```python
# creating a copy of the image
rectangle_image = good_image.copy()

# creating the rectangle
cv2.rectangle(img=rectangle_image, pt1=(180, 370), pt2=(620, 700), color=(255, 0, 0), thickness=4)

# showing the image
plt.imshow(rectangle_image)
```

# 4. Draw a triangle in the middle of the image

```python
# creating a copy of the image
triangle_image = good_image.copy()

# creating the vertices of the polyline
vertices = np.array([[200, 700], [400, 400], [600, 700]])

# reshaping the vertices
points = vertices.reshape((-1, 1, 2))

# creating the traingle
cv2.polylines(img=triangle_image, pts=[points], isClosed=True, color=(0, 0, 255), thickness=8)

# showing the image
plt.imshow(triangle_image)
```

# 5. Try to find a solution on how to fill the triangle

```python
# create a copy of the image
filled_image = good_image.copy()

# creating the vertices of the polyline
vertices = np.array([[200, 700], [400, 400], [600, 700]])

# reshaping the vertices
points = vertices.reshape((-1, 1, 2))

# creating a filled polygon
cv2.fillPoly(img=filled_image, pts=[points], color=(0, 0, 255))

# showing the image
plt.imshow(filled_image)
```

# 6. Create a script that opens the image and draws an empty red circle every time we click the right mouse button

```python
# importing the libraries
import numpy as np
import cv2

# reading the image
image = cv2.imread('../DATA/dog_backpack.png')

# function that draws the circles
def draw_circle(event, x, y, flags, params):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img=image, center=(x, y), radius=70, color=(255, 0, 0), thickness=4)
    
# connecting the function to the image
cv2.namedWindow('dog')
cv2.setMouseCallback('dog', draw_circle)


# showing the image
while True:
    cv2.imshow('dog', image)

    if cv2.waitKey(10) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
